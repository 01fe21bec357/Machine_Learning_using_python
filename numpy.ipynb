```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 10)
y = np.sin(2 * x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y, linewidth=2.0)

# Set custom axis limits and ticks
ax.set_xlim(0, 8)  # Set x-axis limits
ax.set_xticks(np.arange(1,8))  # Set x-axis ticks
ax.set_ylim(0,81)  # Set y-axis limits
ax.set_yticks(np.arange(1,8))  # Set y-axis ticks

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 10)
y = np.sin(2 * x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the initial data
ax.plot(x, y, linewidth=2.0)

# Set custom axis limits and ticks
ax.set_xlim(0, 8)  # Set x-axis limits
ax.set_xticks(np.arange(1, 8))  # Set x-axis ticks
ax.set_ylim(0, 8)  # Set y-axis limits
ax.set_yticks(np.arange(1, 8))  # Set y-axis ticks

# Show the initial plot
plt.show()

# Update the data
x = np.linspace(0, 10, 100)
y = 6 + 3 * np.sin(2 * x)

# Clear the previous plot and plot the new data
ax.clear()
ax.plot(x, y, linewidth=2.0)

# Set custom axis limits and ticks for the updated data
ax.set_xlim(0, 10)  # Set x-axis limits
ax.set_xticks(np.arange(0, 11, 2))  # Set x-axis ticks
ax.set_ylim(0, 12)  # Set y-axis limits
ax.set_yticks(np.arange(0, 13, 2))  # Set y-axis ticks

# Show the updated plot
plt.show()
```
