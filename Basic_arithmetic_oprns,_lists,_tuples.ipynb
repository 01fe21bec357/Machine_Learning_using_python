tuples operations

```python
tuple1=("hubli","belgaum","hubli")
print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
```

```python
print(tuple1[-1])
```

```python
#joining two tuples
tuple6=tuple1+tuple1
print(tuple6)
```

```python
tuple7=tuple6*2
print(tuple7)
```

```python
y=tuple7.index('hubli')
print(y)
```

```python
#how many hublis are there in the tuple
x=tuple7.count('hubli')
print(x)
```
