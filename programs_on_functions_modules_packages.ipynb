modules and packages

args

```python
def add(*args):
    sum = 0
    for i in args:
        sum += i
    return sum

print(add(1, 2))
print(add(1, 2, 3))
```

kwargs

```python
def greet(**kwargs):
  print("hello")
  for key,value in kwargs.items():
    print(value)
greet(name="nag",usn="01fe",branch="ece")
```

```python
def greet(**kwargs):
  print("hello")
  for key,value in kwargs.items():
    print(value,end=" ")
greet(name="nag",usn="01fe",branch="ece")
```

lottery game in python

```python
import random
computer=random.randint(1,9)
userinput=int(input("enter the number between 1 and 9  "))
print(computer)
print(userinput)
if computer>userinput:
  print("computer won")
elif computer==userinput:
  print("TIE")
else:
  print("you won")
```
