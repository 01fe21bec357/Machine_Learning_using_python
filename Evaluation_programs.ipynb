Palindrome

```python
string=input("Enter string : ")

#Declare an empty string variable
revstr=""

#Iterate string with for loop
for i in string:
  revstr=i+revstr
print("Reversed string : ",revstr)

if(string==revstr):
  print("The string is a palindrome.")
else:
  print("The string is not a palindrome.")
```

even odd list

```python
def splitevenodd(A):
   evenlist = []
   oddlist = []
   for i in A:
      if (i % 2 == 0):
         evenlist.append(i)
      else:
         oddlist.append(i)
   print("Even lists:", evenlist)
   print("Odd lists:", oddlist)

# Driver Code
A=list()
n=int(input("Enter the size of the First List ::"))
print("Enter the Element of First  List ::")
for i in range(int(n)):
   k=int(input(""))
   A.append(k)
splitevenodd(A)
```

```python
k = int(input("Enter the number: "))
for i in range(k):
    for j in range(i + 1):
        print("*", end=' ')
    print()  # Print a newline after each row
```

prime number

```python
n = int(input("Enter the value: "))
is_prime = True
divisibility_flags = []

if n <= 1:
    is_prime = False
else:
    for x in range(2, n):
        if n % x == 0:
            is_prime = False
            if x in [3, 7, 9, 11]:
                divisibility_flags.append(x)

if is_prime:
    print("It is a prime number.")
else:
    print("It is not a prime number.")
    if 3 in divisibility_flags:
        print(f"It is divisible by 3.")
    if 7 in divisibility_flags:
        print(f"It is divisible by 7.")
    if 9 in divisibility_flags:
        print(f"It is divisible by 9.")
    if 11 in divisibility_flags:
        print(f"It is divisible by 11.")
```
